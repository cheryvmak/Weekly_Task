# Hands-On Exercise
Task 1: Read a text file, clean the text, and tokenize it into words.

Task 2: Create a Python function that removes stopwords and punctuation from a given string.

Task 3: Use the cleaned and tokenized text to construct a simple prompt for a generative AI model.

```python
import string
import nltk
from nltk.corpus import stopwords

# Download stopwords if not already downloaded
nltk.download('stopwords')

# Step 1: Read, clean, and tokenize a text file
def read_and_tokenize(file_path):
    try:
        with open(file_path, 'r') as file:
            text = file.read()

        # Clean the text (remove punctuation and convert to lowercase)
        clean_text = text.translate(str.maketrans('', '', string.punctuation)).lower()
        words = clean_text.split()
        return words
    except FileNotFoundError:
        return "File not found. Please provide a valid file path."

# Step 2: Remove stopwords and punctuation
def remove_stopwords_and_punctuation(words):
    stop_words = set(stopwords.words("english"))
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)

# Step 3: Generate a simple prompt
def generate_prompt(cleaned_text):
    prompt = f"Based on the following input, provide insights: {cleaned_text}"
    return prompt

# Example workflow
file_path = "example.txt"  # Replace with your file path
tokens = read_and_tokenize(file_path)
cleaned_text = remove_stopwords_and_punctuation(tokens)
prompt = generate_prompt(cleaned_text)

print("Generated Prompt:", prompt)
```

# **Principles of Effective Prompt Engineering**

## **Overview**

Prompt engineering involves designing and crafting inputs (prompts) for AI systems to elicit the most accurate, useful, or creative outputs. The effectiveness of a prompt often determines the quality of the AI's response.


**1. Be Specific:** AI models rely on clear and well-defined instructions to understand the desired task.

  - Ineffective: "Tell me about AI."
  - Effective: "Explain the key components of AI in a single paragraph suitable for beginners."

**2. Provide Context**: AI models perform better when provided with enough information to understand the problem or question fully.
  - Include relevant background information or examples.
  - If the prompt refers to previous content, restate the necessary details.
  - For conversational tasks, maintain continuity by including the conversation history.
- **Examples:**
  - Without Context: "What are the main challenges of AI?"
  - With Context: "In the context of implementing AI in healthcare, what are the main challenges organizations face?"

**3. Be Concise**: While context is important, excessive or unnecessary details can confuse the AI and lead to irrelevant responses.
  - Use simple and direct language.
  - Avoid redundant phrases or overly complex sentences.
  - Focus on the essential information required for the task.
- **Examples:**
  - Ineffective: "Please tell me all the ways I could use AI to improve my small business, and also, can you include examples of companies that have done it before?"
  - Effective: "List three ways AI can improve small business operations, with examples."

**4. Use Explicit Instructions**: Clear instructions help the AI understand the required task and avoid unnecessary assumptions.
  - Specify the tone or style (e.g., formal, casual, creative).
  - State the output structure (e.g., numbered list, bulleted points, essay).
  - Provide word limits or character counts if needed.
- **Examples:**
  - "Write a 200-word summary of the following text."
  - "Generate a list of 5 advantages of AI in education."

**5. Experiment and Iterate**: AI models may not always produce the desired output on the first attempt. Iterative refinement helps improve results.
  - Adjust the level of detail in your prompt.
  - Test different phrasings of the same question.
  - Evaluate the output and refine the prompt based on what worked and what didn’t.
  - Iteration 1: "What are the benefits of renewable energy?"
  - Iteration 2: "List three economic benefits of renewable energy in bullet points."

**6. Provide Examples (Few-Shot Learning)**: Including examples in the prompt helps the model understand the desired format or type of response.
  - Provide one or more sample inputs and outputs.
  - Use examples that closely match the complexity of the desired task.
- **Examples:**
  - Who is the president of USA, Mexico, and Australia? For example, Canada: Yetunde Folajimi; France: Emmanuel Macron; Nigeria: Ade Tinubu
CoPilot output:

Here are the current leaders of the USA, Mexico, and Australia:

United States: President Joe Biden

Mexico: President Claudia Sheinbaum

Australia: Prime Minister Anthony Albanese



**7. Avoid Common Pitfalls**: Ineffective prompts can lead to irrelevant or incoherent responses.
- **Common Pitfalls to Avoid:**
  - **Overloading the Prompt:** Including too much information can confuse the model.
  - **Vague Questions:** Avoid unclear or overly broad instructions.
  - **Neglecting the Model’s Capabilities:** Ensure the task aligns with the AI model's strengths.
- **Examples:**
  - Ineffective: "Write a poem and summarize this research paper."
  - Effective: Split into two prompts:
    1. "Write a short poem about AI."
    2. "Summarize this research paper in 100 words."

**8. Leverage Role-Playing**: Assigning a specific role to the AI helps tailor the response to the desired perspective or tone.
- **Tips for Role-Playing:**
  - Define the role explicitly (e.g., "You are a teacher," "You are a data analyst").
  - Frame the task in terms of the assigned role.
- **Examples:**
  - "You are a teacher explaining AI to high school students. Write a simple explanation of what AI is."
  - "As a data analyst, create a report summarizing trends in AI adoption."

**9. Use Chains for Complex Tasks**:Breaking a complex task into smaller, sequential steps improves the quality and relevance of the output.
- **Tips for Chaining Prompts:**
  - Use the output of one prompt as the input for the next.
  - Specify the task in each step.
- **Examples:**
  - Step 1: "Generate a list of key points about climate change."
  - Step 2: "Using the key points, write a summary suitable for a news article."




# Few-Shot and Zero-Shot Learning
## Few-Shot Learning
- Providing the model with a few examples within the prompt to guide its behavior.
- Used for for more complex or ambiguous tasks.
- Useful for tasks where the model needs more context to produce accurate outputs.
- Example:
 - Translate the following sentences from English to French:
   - English: "How are you?"
     - French: "Comment ça va?"
   - English: "What is your name?"
     - French: "Comment vous appelez-vous?"
  - English: "Where is the library?"
    - French:
      - Result: The model continues the pattern and outputs: "Où est la bibliothèque?"

## Zero-Shot Learning
- Instructing the model to perform a task without providing any examples.
- Relies on a well-crafted prompt to specify the task.
- For straightforward tasks or when brevity is key.
- Example:
  - Provide three unique names for a pet dog.
  - Result: The model generates names like: "Buddy, Luna, Max."

#Chained Prompts:
Breaking a complex task into smaller, step-by-step prompts.
- Useful for tasks that require logical progression or multiple outputs.
- The model's understanding builds on previous outputs.
- Requires precise phrasing to maintain coherence across prompts.
- Example:
  - Prompt 1: "Generate a list of three potential topics for a blog on AI ethics."
  - Model Output: "1. Bias in AI; 2. Data Privacy; 3. AI in Decision-Making."
  - Prompt 2: "Expand on the first topic, 'Bias in AI,' and list possible solutions."
  - Model Output: "Bias in AI can be addressed through diverse datasets, fairness metrics, and human oversight."


# Dynamic Prompt Generation
- Using scripts to programmatically generate prompts based on user input or data.
- Saves time when dealing with repetitive tasks or structured data.
-Use cases:
  - Automatically creating prompts for large datasets.
  - Personalizing responses in applications like chatbots or virtual assistants.





```python
#Dynamic prompt example
templates = [
    "Write a summary of the following text: {}",
    "Explain the significance of: {}",
    "Generate three questions about: {}"
]

data_points = ["Artificial Intelligence", "Climate Change", "Quantum Computing"]

for data in data_points:
    for template in templates:
        prompt = template.format(data)
        print(prompt)
```

## Role Playing
- Instructs the model to act as a specific role to produce more tailored outputs.
- Guides the model to adopt a tone, style, or perspective suited to the task.
- Makes responses more relatable and audience-appropriate.
- Example:
  - Prompt: "You are a teacher explaining quantum physics to a high school student. Write a simple explanation."
  - Output: "Quantum physics studies very tiny particles and how they behave differently than we expect..."

## Instruction-Based Prompts
- Providing explicit, step-by-step instructions for the model to follow.
- Example:
  - Step 1: Analyze the following text and identify the main idea.
  - Step 2: Provide two supporting details from the text.
  - Step 3: Summarize your findings in one sentence.
- Output:
  - Main idea: "AI can transform healthcare."
  - Supporting details: "AI aids in diagnosis. AI optimizes treatment."
  - Summary: "AI has the potential to revolutionize healthcare through diagnosis and treatment optimization."
- Benefits:
  - Produces structured, predictable outputs.
  - Especially useful for academic or professional tasks.


## Iterative Refinement
- Experimenting with multiple versions of a prompt to achieve better results.
- Steps:
  - Start with a basic prompt.
  - Analyze the model’s output for clarity, relevance, and completeness.
  - Refine the prompt with additional context, instructions, or examples.
- Example:
 - Initial Prompt: "Write a poem about the moon."
 - Output: Generic poem.
 - Refined Prompt: "Write a romantic poem about the moon, focusing on its impact on lovers."
 - Output: More specific, engaging poem.

```python
#Use Transformer
from transformers import pipeline
```

```python
# Example 1: Text Summarization

# Load the text summarization model
summarizer = pipeline('summarization')

# Input text
text = """
Artificial Intelligence (AI) is revolutionizing various industries such as healthcare, education, and finance.
With advancements in machine learning and natural language processing, AI is becoming more accessible and powerful.
"""

# Generate the summary
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)
print("Original Text:", text)
print("Summary:", summary[0]['summary_text'])
```

```python
#Example 2: Sentiment Analysis
#from transformers import pipeline

# Load the sentiment analysis model
sentiment_analyzer = pipeline('sentiment-analysis')

# Sample texts
texts = [
    "I love the new AI features in this app! They're amazing.",
    "This product is the worst thing I’ve ever bought."
]

# Analyze sentiment
for text in texts:
    result = sentiment_analyzer(text)
    print(f"Text: {text}")
    print(f"Sentiment: {result[0]['label']}, Confidence: {result[0]['score']:.2f}\n")
```

```python
# Example 3: Question Answering

from transformers import pipeline

# Load the question-answering model
qa_model = pipeline('question-answering', model='distilbert-base-cased-distilled-squad', device=0)
# Context and question
context = """
Python is a versatile programming language widely used in data science, web development, and artificial intelligence.
It is known for its simplicity and readability.
"""
question = "What is Python known for?"

# Generate the answer
answer = qa_model(question=question, context=context)
print(f"Question: {question}")
print(f"Answer: {answer['answer']}")
```
