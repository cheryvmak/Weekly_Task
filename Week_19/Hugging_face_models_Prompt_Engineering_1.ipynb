# **Case Studies for LLMs**
## Sequence-to-Sequence (Seq2Seq) Tasks and Transformer Architecture
Seq2Seq is a supervised learning problem where the input is a sequence, and the output is another sequence of potentially different lengths.

## Examples:
- Machine Translation: Translating a sentence from one language to another.
- Text Summarization: Generating a concise summary of a long document.
- Speech-to-Text: Converting spoken words into written text.
- Dialogue Systems: Generating coherent responses in chatbots.

## Traditional Seq2Seq Models
1. Encoder-Decoder Framework:
  - Encoder: Processes the input sequence and encodes it into a fixed-length context vector.
  - Decoder: Generates the output sequence from the context vector.

## Challenges:
  - Fixed-length Bottleneck: Encoding the entire input sequence into a single vector limits the model's ability to handle long sequences.
  - Vanishing Gradients: RNN-based models like LSTMs and GRUs struggle with long-range dependencies.
  - Sequential Processing: Traditional models process sequences step-by-step, making training and inference slow.

# Transformers: Revolutionizing Seq2Seq Tasks
The Transformer architecture, introduced in the 2017 paper "Attention is All You Need", has significantly improved the efficiency and performance of Seq2Seq tasks.

Key Features of Transformer Architecture
1. Attention Mechanism:
  - Replaces sequential processing with parallelizable attention computations.
  - Self-Attention allows the model to focus on relevant parts of the input sequence at each position.
  - Scaled Dot-Product Attention computes relevance scores between all pairs of tokens.

2. Encoder-Decoder Structure:
  - Encoder: Converts the input sequence into contextualized embeddings using multiple attention layers.
  - Decoder: Uses these embeddings and attention to generate the output sequence step-by-step.

3. Parallelism:
  - Transformers process entire sequences at once, unlike RNNs, which are sequential. This parallelization makes training significantly faster.
Positional Encoding:
  - Since transformers lack inherent sequential order (unlike RNNs), positional encodings are added to token embeddings to capture sequence order.
Scalability:
  - Transformers handle much larger datasets and sequences effectively, thanks to their multi-head attention and efficient handling of long-range dependencies.


Advantages of Transformers for Seq2Seq Tasks
1. Efficiency:
  - Parallel processing of tokens reduces training and inference time.
  - Can handle longer sequences without significant performance degradation.
2. Handling Long-Range Dependencies:
  - Attention mechanisms allow the model to focus on relevant parts of the sequence, regardless of their position.
3. Modularity:
  - Easy to adapt the architecture for specific tasks (e.g., GPT for text generation, BERT for classification).
4. State-of-the-Art Results:
  - Achieved breakthroughs in tasks like translation (e.g., Google Translate), summarization, and dialogue generation.


# Application Examples

````python
from datasets import load_dataset

# Load IMDB dataset as an example
dataset = load_dataset("imdb")

# Split dataset into training and validation sets
train_data = dataset["train"]
val_data = dataset["test"]

# Preprocess text using a tokenizer
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def preprocess_function(examples):
    return tokenizer(examples["text"], truncation=True, padding=True)

train_data = train_data.map(preprocess_function, batched=True)
val_data = val_data.map(preprocess_function, batched=True)
````

````python
#Load a Pre-trained Model
from transformers import AutoModelForSequenceClassification

# Load BERT with a classification head
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)  # 2 categories
````

````python
#Define Training Arguments

from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
)
````

````python
#Create a Trainer
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_data,
    eval_dataset=val_data,
)
````

````python
#Train the Model
trainer.train()
````

````python
# Finetuning with reduced dataset
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from transformers import EarlyStoppingCallback
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# Load and preprocess dataset
dataset = load_dataset("imdb")
small_train_data = dataset["train"].shuffle(seed=42).select(range(1000))
small_val_data = dataset["test"].shuffle(seed=42).select(range(500))

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

def preprocess_function(examples):
    return tokenizer(examples["text"], truncation=True, padding=True)

train_data = small_train_data.map(preprocess_function, batched=True)
val_data = small_val_data.map(preprocess_function, batched=True)

# Load pre-trained model
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)

# Define metrics function
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = predictions.argmax(axis=1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, predictions, average="binary")
    acc = accuracy_score(labels, predictions)
    return {"accuracy": acc, "f1": f1, "precision": precision, "recall": recall}

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",  # Deprecated, but you can use it for now; replace with eval_strategy in future
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    learning_rate=2e-5,
    load_best_model_at_end=True,  # Required for EarlyStoppingCallback
    metric_for_best_model="accuracy",  # Specify the metric to monitor
    logging_dir="./logs",
    save_strategy="epoch",  # Save checkpoints every epoch
    fp16=True,  # Mixed precision for faster training
)

# Create Trainer with early stopping
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_data,
    eval_dataset=val_data,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)],
)

# Train and evaluate
trainer.train()
````

````python
#Evaluate the Fine-Tuned Model
trainer.evaluate()
````

````python
# Save and use the finetuned model
model.save_pretrained("./fine_tuned_model")
tokenizer.save_pretrained("./fine_tuned_model")

# Load the saved model for inference
from transformers import pipeline

classifier = pipeline("text-classification", model="./fine_tuned_model")
result = classifier("The movie was awesome!")
print(result)
````

# Token Classification with Hugging Face

Token classification is a Natural Language Processing (NLP) task that assigns a label to each token (word or subword) in a sequence. This is commonly used in tasks such as **Named Entity Recognition (NER)**, **Part-of-Speech (POS) tagging**, and **chunking**.



#### **1. Common Use Cases of Token Classification**
1. **Named Entity Recognition (NER)**:
   - Identifies named entities like people, organizations, locations, dates, and more.
   - Example:
     - Input: "Elon Musk founded SpaceX in 2002."
     - Output:
       - `Elon Musk → PERSON`
       - `SpaceX → ORG`
       - `2002 → DATE`

2. **Part-of-Speech (POS) Tagging**:
   - Assigns grammatical roles to each word in a sentence (e.g., noun, verb, adjective).
   - Example:
     - Input: "The cat sat on the mat."
     - Output: `The → DET`, `cat → NOUN`, `sat → VERB`.

3. **Chunking**:
   - Groups tokens into meaningful phrases like noun or verb phrases.
   - Example:
     - Input: "The quick brown fox."
     - Output: `[The quick brown fox] → NP (Noun Phrase)`.



#### Using Hugging Face for Token Classification

# Applications of Token Classification
- Named Entity Recognition: Extract entities like names, dates, and locations from legal or financial documents.
- Information Extraction: Identify specific phrases or patterns in customer feedback or reviews.
- Part-of-Speech Tagging: Analyze sentence structure in linguistics or grammar-checking applications.
- Custom Entity Detection: Use for domain-specific tasks like detecting chemical compounds or medical conditions.


````python
from transformers import pipeline

# Specify the model explicitly
model_name = "dbmdz/bert-large-cased-finetuned-conll03-english"
revision = "4c53496"  # Specific model revision

# Load token classification pipeline with specified model and revision
token_classifier = pipeline("token-classification", model=model_name, revision=revision, device=0)

# Perform token classification
text = "Mike Tyson founded SpaceX, headquartered in Hawthorne, California, in 2002."
results = token_classifier(text)

# Print results
for entity in results:
    label = entity.get('entity_group', entity.get('entity'))
    print(f"Entity: {entity['word']}, Label: {label}, Confidence: {entity['score']:.2f}")
````

# **Text Generation in NLP**

Text generation is an NLP task where a model generates coherent and contextually relevant text based on a given input prompt. It involves predicting the next word(s) in a sequence until a desired output length or stopping condition is met.

## **Common Use Cases**
1. **Creative Writing**:
   - Generating poems, stories, or song lyrics.
2. **Dialogue Systems**:
   - Creating chatbot responses.
3. **Text Autocompletion**:
   - Suggesting the next word or phrase (e.g., Gmail Smart Compose).
4. **Code Generation**:
   - Writing code snippets based on prompts.
5. **Content Creation**:
   - Generating articles, social media posts, or product descriptions.


## **How Does Text Generation Models Work**
- The model generates text one token at a time, using:
  - **Probabilities**: Based on training data, it predicts the most likely next token.
  - **Beam Search / Sampling**: Techniques to improve fluency or creativity of generated text.

# Customizing Parameters
Hugging Face allows fine control over the generation process through several parameters:
- max_length: Maximum number of tokens in the output.
- num_return_sequences: Number of different outputs to generate.
- temperature: Controls randomness; lower values (e.g., 0.7) result in - more focused text, while higher values (e.g., 1.5) increase creativity.
- top_k: Limits sampling to the top K tokens (helps with coherence).
- top_p: Uses nucleus sampling to consider tokens with a cumulative probability ≥ p.

````python
#EXample:
output = generator(
    "The future of technology is",
    max_length=30,
    temperature=0.8,
    top_k=50,
    top_p=0.9,
    num_return_sequences=3
)

for i, text in enumerate(output):
    print(f"Generated text {i+1}: {text['generated_text']}")
````

## 2. Import Necessary Libraries
torch: Provides tensor operations, which are required to handle inputs and outputs for the model.

transformers: From Hugging Face, this library is used to load pre-trained models like DialoGPT and manage tokenization.

````python
#Import Libraries
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
````

## 3. Load the Model and Tokenizer
microsoft/DialoGPT-medium: A pre-trained conversational model from Microsoft, based on GPT-2 and fine-tuned on dialogue datasets (e.g., Reddit).

AutoTokenizer: Converts text into tokens (IDs) that the model can process and converts the model's output back into human-readable text.

AutoModelForCausalLM: Loads the conversational model that uses causal language modeling (predicts the next token based on previous ones).



````python
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
````


## 3. Setting the Padding Token

pad_token: Padding tokens are used when input sequences of varying lengths are batched together.
Since DialoGPT doesn’t have a dedicated padding token, it uses the end-of-sentence (eos_token) as a placeholder


````python
tokenizer.pad_token = tokenizer.eos_token
````


## 4. The chat_with_bot Function

This function generates a response from the chatbot.

### Function Parameters
user_input: The text provided by the user.

chat_history_ids: A tensor containing the conversation history (previous exchanges). Defaults to None for the first turn.

### Adding a Guiding Prompt

Prompt: A guiding instruction for the chatbot to behave in a specific way (e.g., polite and helpful). This helps improve the quality and relevance of responses.

tokenizer.encode: Converts the prompt and user input into token IDs, which the model processes.

+ tokenizer.eos_token: Appends the eos_token to mark the end of the input text.

return_tensors="pt": Ensures the output is in PyTorch tensor format.

prompt = f"You are a professional and polite assistant. Respond helpfully: {user_input}"

input_ids = tokenizer.encode(prompt + tokenizer.eos_token, return_tensors="pt")



### Combining Conversation History
torch.cat: Concatenates the previous conversation history (chat_history_ids) with the current user input (input_ids), ensuring the model has the full context of the conversation.

If chat_history_ids is None (e.g., the first message), only the input_ids are used.


````python
bot_input_ids = torch.cat([chat_history_ids, input_ids], dim=-1) if chat_history_ids is not None else input_ids
````

### Generating a Response

model.generate: Generates a response from the chatbot.

bot_input_ids: Includes the full conversation history and current user input.

max_length=1000: Sets the maximum length of the output, ensuring the conversation doesn't grow too long.

pad_token_id=tokenizer.pad_token_id: Specifies the padding token to handle variable-length inputs.

attention_mask=torch.ones_like(bot_input_ids): Ensures all tokens in bot_input_ids are attended to.

do_sample=True: Enables sampling for diverse and creative responses.


temperature=0.7: Introduces randomness; lower values make responses more deterministic, while higher values make them more creative.

top_p=0.9: Enables nucleus sampling, which selects tokens from the top 90% of cumulative probability.

top_k=50: Restricts sampling to the top 50 tokens, improving coherence.

repetition_penalty=1.2: Penalizes repeated phrases or tokens to reduce repetitive responses.

### Decoding the Response

chat_history_ids[:, bot_input_ids.shape[-1]:]: Extracts only the newly generated part of the conversation, ignoring the input tokens.

tokenizer.decode: Converts the token IDs back into human-readable text.

skip_special_tokens=True: Removes special tokens like <eos> or <pad> from the output.


response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)


### Return Values
response: The chatbot's reply as a string.

chat_history_ids: Updated conversation history, including the latest user input and bot response.



````python
# Chat function
def chat_with_bot(user_input, chat_history_ids=None):
    # Add a guiding prompt to improve behavior
    prompt = f"You are a professional and polite assistant. Respond helpfully: {user_input}"
    input_ids = tokenizer.encode(prompt + tokenizer.eos_token, return_tensors="pt")
    bot_input_ids = torch.cat([chat_history_ids, input_ids], dim=-1) if chat_history_ids is not None else input_ids

    # Add attention mask and use sampling
    attention_mask = torch.ones_like(bot_input_ids)
    chat_history_ids = model.generate(
        bot_input_ids,
        max_length=1000,
        pad_token_id=tokenizer.pad_token_id,
        attention_mask=attention_mask,
        do_sample=True,        # Enable sampling
        temperature=0.7,       # Add randomness
        top_p=0.9,             # Use nucleus sampling
        top_k=50,              # Restrict to top 50 tokens
        repetition_penalty=1.2 # Penalize repetition
    )

    # Decode the response
    response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)
    return response, chat_history_ids
````

### Conversation Loop


````python

print("Chatbot: Hi! How can I help you?")
chat_history_ids = None

while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye!")
        break
    if user_input.lower() == "reset":
        chat_history_ids = None
        print("Chatbot: Conversation history reset.")
        continue

    response, chat_history_ids = chat_with_bot(user_input, chat_history_ids)
    print(f"Chatbot: {response}")
````
